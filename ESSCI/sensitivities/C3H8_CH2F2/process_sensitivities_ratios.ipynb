```python
#subtraction sensitivity calc for ANL_Brown5

# Dr. West: "(X_perturbed - X_ref)/X_ref / (delta) where X=(tau@4%/tau@0.1%) " 


def get_x_ref(temp, current): 


    dict_key_4 = f'{temp}_4.0'
    dict_key_pt1 = f'{temp}_0.1'


    if current==False: 
        initial_ANL_Brown5 = {'1818.18_0': 0.009214139279580, 
                               '1818.18_4.0': 0.10725536943699, 
                               '1467.18_0': 0.29328656790373, 
                               '1467.18_4.0': 0.9658373777952,
                               '1292.51_0': 4.072755074569, 
                               '1292.51_4.0': 2.939843872732,
                               '1292.51_0.1': 3.3073233557855,
                               '1467.18_0.1': 0.3050323529151,
                               '1818.18_0.1': 0.01425340830212
                            }
    if current==True: 
        initial_ANL_Brown5 = {'1818.18_0': 0.011791662661626, 
                       '1818.18_4.0': 0.1020623719138, 
                       '1467.18_0': 0.2569873072756, 
                       '1467.18_4.0': 1.4223707768067,
                       '1292.51_0': 3.627222759753, 
                       '1292.51_4.0': 8.219724270751,
                       '1292.51_0.1': 3.7754333390007,
                       '1467.18_0.1': 0.29174646456929,
                       '1818.18_0.1': 0.015110066353340 #taking off three decimals
                    }


    #now divide the 4% by the 0.1%
    tau_i_4 = initial_ANL_Brown5[dict_key_4]
    tau_i_pt1 = initial_ANL_Brown5[dict_key_pt1]
    
    print(tau_i_4 ,tau_i_pt1, rxn_index)  
    
    x_ref = tau_i_4 / tau_i_pt1
    return x_ref

def get_x_perturbed(temp, rxn_index):
    
    if temp==1292.51: 
        tau_new_4 = ign_delays_4_1292[rxn_index]
        tau_new_pt1 = ign_delays_pt1_1292[rxn_index]
    if temp==1467.18: 
        tau_new_4 = ign_delays_4_1467[rxn_index]
        tau_new_pt1 = ign_delays_pt1_1467[rxn_index]
    if temp==1818.18: 
        tau_new_4 = ign_delays_4_1818[rxn_index]
        tau_new_pt1 = ign_delays_pt1_1818[rxn_index]
   
    print(tau_new_4 ,tau_new_pt1, rxn_index) 

    x_perturbed = tau_new_4/tau_new_pt1
    return x_perturbed
    
def get_sensitivity(temp, dk, rxn_index, current):
    
    x_ref = get_x_ref(temp, current) 
    x_perturbed = get_x_perturbed(temp, rxn_index)
    S = (1/x_ref)*((x_perturbed - x_ref)/dk)
    return S
    
```

```python
def just_give_me_indices(x, reactants=False, products=False):
    try: 
        if reactants: 
            index_list = sorted([int(re.search('\(([0-9]+)\)', x).group(1)) for x in list(pre_gas.reaction(x).reactants.keys())])
        if products:
            index_list = sorted([int(re.search('\(([0-9]+)\)', x).group(1)) for x in list(pre_gas.reaction(x).products.keys())])
    except AttributeError as e: 
        return None
    return index_list
```
