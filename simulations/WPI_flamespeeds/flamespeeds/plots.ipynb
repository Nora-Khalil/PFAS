```python
#ANL_Brown5
output_files = [file for file in os.listdir('./blend/ANL_Brown5') if re.search('output_fc', file)]
results = {}
for file in output_files:
    print(file)
    path_ = f'./blend/ANL_Brown5/{file}'
    with open(path_, 'r') as f: 
        lines = f.readlines()
        if lines != []:
            vf = float(lines[-3])
            speed = float(lines[-1])*100
            results[vf]=speed
        
vfs = sorted(results.keys())        
speeds = [results[vf] for vf in sorted(results.keys())]

plt.plot(vfs, speeds, '-', c='blue', label = 'ANL_Brown5')   

plt.title('ANL_Brown5 | Blend')
plt.xlabel('CH2F2 %')
plt.ylabel('flame speeds (cm/s)')
plt.legend()
        

```
