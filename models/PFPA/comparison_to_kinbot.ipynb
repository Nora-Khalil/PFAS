```python
unimolecular_rxns_with_PFPA = []
PFPA=spec_RMG_edge[2]
for rxn in reac_RMG_edge:
    for reactant in rxn.reactants:
        if len(rxn.reactants)==1:
            if reactant==PFPA:
                unimolecular_rxns_with_PFPA.append(rxn)
    for product in rxn.products:
        if len(rxn.products)==1:
            if product==PFPA:
                unimolecular_rxns_with_PFPA.append(rxn)
```

```python
for rxn in unimolecular_rxns_with_PFPA:
    print(rxn)
    indices = []
    for reactant in rxn.reactants: 
        target = reactant.index
        for ind, spec in enumerate(spec_RMG_edge):
            if spec.index==target:
                reactant_index=ind
                indices.append(reactant_index)
            
    for product in rxn.products:
        target = product.index
        for ind, spec in enumerate(spec_RMG_edge):
            if spec.index==target:
                product_index=ind
                indices.append(product_index)
    print(indices)
    print('\n','\n')
    
        
```
