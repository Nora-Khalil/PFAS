```python
index=-1 #start at 0

final_string = ''

for spc in all_species:
    print(spc.name)
    
    #increment the index 
    index+=1
    
    #get the label
    label = spc.name
    
    #get the adjacency list
    value = species_structure_dict[spc.name]
    #print(value)
    if type(value)==str:
        SMILES = value
        adjacency_list = Molecule(smiles=SMILES).to_adjacency_list()
    if type(value)==list: 
        SMILES = value[0]
        adjacency_list = value[1]

    
    #now for the thermo
    max_temp = spc.thermo.max_temp
    min_temp = spc.thermo.min_temp
    middle_temp = spc.thermo.coeffs[0]
    first_T_region = list(spc.thermo.coeffs[-7:])
    second_T_region = list(spc.thermo.coeffs[1:8])
    cp0 = spc.thermo.cp(0)/1000 #cantera has it in J/kmol/K
    

    note = spc.thermo.input_data['note']
    string = f'''

entry(
    index = {index},
    label = "{label}",
    molecule = 
"""
{adjacency_list}""",
    thermo = NASA(
        polynomials = [
            NASAPolynomial(coeffs={first_T_region}, Tmin=({min_temp},'K'), Tmax=({middle_temp},'K')),
            NASAPolynomial(coeffs={second_T_region}, Tmin=({middle_temp},'K'), Tmax=({max_temp},'K')),
        ],
        Tmin = ({min_temp},'K'),
        Tmax = ({max_temp},'K'),
        Cp0 = ({cp0},'J/(mol*K)'),
    ),
    reference = {Reference},
    referenceType = "Theory",
    shortDesc = "{shortDesc}",
    longDesc = """{longDesc}""",
)
'''
    final_string+=string
```
